```python
# Experiment
# Chems->cid
# distance, centroids,cluster
# store 
#[(i.eid,i.name,len(i.chems)) for i in HtsExp.objects]
#ChemClusterNode.drop_collection()
#HtsExpLink.drop_collection()
for Ex1 in HtsExp.objects:
    print '>',Ex1.name
    CID = set([i.cid for i in Ex1.chems if i.cid])
    print ' ', len(CID),'chemicals'
    CC=ChemCluster()
    CC.load_chem_ids(CID)
    CC.clust_chems(fp='maccs2')
    CC.find_centroids(k_cent=30,km_ninit=10)
    CC.clust_centroids(metric='euclidean',link='complete')
    name='chem-clust-3-'+Ex1.eid
    Clst1 = storeChemCluster(CC._Cent_G,meth=CH_cent_clus,name=name,desc='Clustering chemicals for HCS experiment. '+Ex1.name)
    Link=HtsExpLink(exp=Ex1,link_db='txp-db',link_id=Clst1.id,link_cls='ChemClusterNode',tags=['cluster','chem'])
    Link.save()
```

```python
def cleanStr(x):
    if x[-1]=='"': x=x[:-1]
    if x[0]=='"': x=x[1:]
    return x

for C in Chemical.objects(tags='toxcast'):
    print '>',C.name,C.casrn
    Ents = txp.Entity.objects(Q(name=C.name,src='DSSTOX') | Q(syn__in=C.casrn,src='DSSTOX'))
    E = None
    if not Ents:
        print '  Not in DSSTOX!'
        X = sorted([(i,len(i.syn)) for i in txp.Entity.objects(Q(name=C.name) | Q(syn__in=C.casrn))],key=lambda z:z[1],reverse=True)
        if X: E = X[0][0]
    elif len(Ents)>0:
        print '  Multiple Entries'
        E = Ents.first()
    
    Syn = set(C.casrn + C.syn) 
    if E: Syn.update([cleanStr(s) for s in E.syn if len(s)<25])
    Syn = list(Syn)
    if E: print ' ',E.name,E.src,','.join(Syn)[:40]
    if Syn: 
        C.syn = Syn
        C.save()
```
